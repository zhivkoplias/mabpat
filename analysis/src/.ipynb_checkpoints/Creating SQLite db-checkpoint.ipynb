```python
def get_patent_num(jstr):
    import re
    jstr = str(jstr)
    pats = re.findall('(: .+?(?=(\s\d*-\w\d*)))', jstr)
    if not pats:
        #print(pats)
        return None
    else:
        pats = pats[0]
        joiner = "".join
        return "".join(list(filter(None, [joiner(words) for words in pats])))[2::].replace(" ", "").replace("-", "")
```

## Parse accession number

```python
DB = '/home/erikz/data/patents_db/genepat_Db_V4_with_accession.sqlite'
FILEDIR = '/home/erikz/data/patents_db/genbank_data_V4/'
```

```python
all_genefiles = sorted([f for f in os.listdir(FILEDIR) if f.endswith('.seq.gz')])
```

```python
all_genefiles = all_genefiles
```

```python
def get_journal(seq):
    
    try:
        ix_j = seq.index('JOURNAL')
        ix_f = seq.index('FEATURES')
        
        journal = seq[ix_j:ix_f].strip()
        
    except:
        return None # Returns None if substring doesnt exist
    
    return journal
```

```python
def get_origin(seq):
    
    try:
        ix_j = seq.index('ORIGIN ')
        
        origin = seq[ix_j:].split('\n')
        origin = ''.join([line[10:].replace(' ','') for line in origin[1:]])
        
    except:
        return None # Returns None if substring doesnt exist
    
    return origin
```

```python
## Create table for Db
conn = sqlite3.connect(DB)
cur = conn.cursor()

## SQL used to create db:
cur.execute('''CREATE TABLE "patents_accession" ("id" INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL UNIQUE, "accession" TEXT, "locus" TEXT, "organism" TEXT, "definition" TEXT, "journal" TEXT, "filename" TEXT, "origin" TEXT)''')    

conn.commit()
conn.close()
```

```python
## For each file - insert it´s contents to db 
start1 = timer()
for genefile in all_genefiles: 

    print('Processing file...: %s' % genefile)
    
    # Read zipped file while decoding from bytes to string
    f = gzip.open(FILEDIR + genefile,'rb')
    ff = f.read().decode()

    # Split up file into separate patent chunks
    sequences = ff.split('\n//\n') # Yes, it gets the first one correct too! 
    # ff.count('//') # nr of separators... should be -1 from len of sequences

    # Grab out the data we want 
    ll = []

    for seq in sequences:
        lines = seq.splitlines()[:]  # ORGANISM, JOURNAL, DEFINITION
        lines = [line.strip() for line in lines]

        ll.append( [
            ''.join([line for line in lines if line.startswith('ACCESSION')]),
            ''.join([line for line in lines if line.startswith('LOCUS')]), 
            ''.join([line for line in lines if line.startswith('ORGANISM')]), 
            ''.join([line for line in lines if line.startswith('DEFINITION')]),
            #''.join([line for line in lines if line.startswith('JOURNAL')]),  
            get_journal(seq), # Get out JOURNAL section using helper method above 
            genefile,
            get_origin(seq)
                ]
        )

    if ll[-1][0] == '': # if the last one is empty (probably always...), then remove it
        ll = ll[:-1]
        print('Last post removed')
        
    # Add data to db
    # tbl: patent 
    conn = sqlite3.connect(DB)
    cur = conn.cursor()
    
    cur.executemany('insert into patents_accession(accession, locus, organism, definition, journal, filename, origin) values (?,?,?,?,?,?,?)', ll)
    conn.commit()
    conn.close()
    
end1 = timer()
print('The total runtime is',(end1-start1),' seconds.')
```
