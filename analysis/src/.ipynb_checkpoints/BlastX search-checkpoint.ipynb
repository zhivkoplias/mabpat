### Download trembl

```python
import requests, sys, json

WEBSITE_API = "https://rest.uniprot.org"

# Helper function to download data
def get_url(url, **kwargs):
    response = requests.get(url, **kwargs);

    #print('DONE')
    if not response.ok:
        print(response.text)
        response.raise_for_status()
        sys.exit()

    return response
```

```python
def get_pd_df_uniprot(species):
    
    import sys
    from io import StringIO
    
    lst_with_results = []
    
    r = get_url(f"https://rest.uniprot.org/uniprotkb/search?fields=accession%2Cid%2Creviewed%2Corganism_name%2Clineage%2Clineage_ids&format=tsv&query=%28{species}+NOT+%28taxonomy_id%3A9606%29%29+AND+%28reviewed%3Afalse%29&size=500")
        
    TESTDATA = StringIO(r.text)
    temp_df_uniprot = pd.read_csv(TESTDATA, sep="\t")
    
    temp_df_uniprot.to_csv(f"../../trembl_db/candidate_ids/{species}.txt", sep='\t', encoding='utf-8')
    
    return(type(temp_df_uniprot))
```

### Download fasta

```python
def get_fasta_uniprot(species, ids_list_original):
    
    import sys
    from io import StringIO
    from Bio import SeqIO
    
    chunk_size = 200000
    if True:
#     if len(ids_list_original)<=chunk_size:
    
        r = get_url(f"https://rest.uniprot.org/uniprotkb/search?format=fasta&query={species}&facets&reviewed%3Afalsesize=50000")
        
        records = SeqIO.parse(StringIO(r.text), "fasta")
        
        with open(f"../../trembl_db/candidate_ids/{species}.faa", "w") as handle:
            SeqIO.write(records, handle, "fasta")
    else:
        list_with_records = []
        num_chunks = len(ids_list_original) // chunk_size + 1
        for i in range(num_chunks):
            new_query = ids_list_original[i*chunk_size:(i+1)*chunk_size]
            new_query = '+OR+'.join(new_query)
            #print(new_query)
            r = get_url(f"https://rest.uniprot.org/uniprotkb/search?format=fasta&query={new_query}")
        
            records = SeqIO.to_dict(SeqIO.parse(StringIO(r.text), "fasta"))
            #print(len(records.values()))
            #print(records)
            list_with_records.append(records)        
        for records in list_with_records:
            with open(f"../../trembl_db/candidate_ids/{species}.faa", "a+") as handle:
                SeqIO.write(records.values(), handle, "fasta")
                handle.close()
    
    print(f"{species} completed")
    return
```

```python
for species_name in taxonomy_list[0:5]:
    get_fasta_uniprot(species_name, ['g'])
```
